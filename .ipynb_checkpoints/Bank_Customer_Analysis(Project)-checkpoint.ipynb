```python
# util data distribution visualization functions

def draw_hist(df, col):
    plt.hist(df[col], color='green', edgecolor='black')
    plt.title(f'Histogram plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    
def draw_count_plot(df, col):
    sb.countplot(y=col, data=df, palette='Greys_r')
    plt.title(f'Countplot of {col}')
    plt.show()
```

## Question 1
1. Demographic Analysis
   -	What is the distribution by geography, gender, age?
   -	Do demographics affect credit score, balance, or salary?


###  - What is the distribution by geography, gender, age?

```python
# Distribution by gender & geography
sb.countplot(x='Gender', data=comb_df, color='green', edgecolor='black')
plt.title(f' Count plot of Customer Distribution by Gender')
plt.show()

sb.countplot(x='Geography',data=comb_df, color='green', edgecolor='black')
plt.title(f'Count plot of Customer Distribution by Geography')
plt.show()

# Age distribution
sb.histplot(comb_df['Age'], color='green', edgecolor='black', bins=20, kde=True)
plt.title(f' Histogram plot of Age Distribution')
plt.show()
```

#### We have more male cutomers than female
#### By Geography, france has the highest customers
#### By Age distributions, most customers are close to 40years of Age

### - Do demographics affect credit score, balance, or salary?

```python
# Mean credit score by gender
sb.barplot(x='Gender', y='CreditScore', data=comb_df, estimator='mean')
plt.title("Average Credit Score by Gender")
plt.show()

# Mean balance by geography
sb.barplot(x='Geography', y='Balance', data=comb_df, estimator='mean')
plt.title("Average Balance by Geography")
plt.show()
```

```python
comb_df.groupby('Geography')['EstimatedSalary'].mean().sort_values(ascending=False)
```

```python
comb_df.groupby('Gender')['EstimatedSalary'].mean().sort_values(ascending=False)
```

```python
comb_df.groupby('Geography')['Balance'].mean().sort_values(ascending=False)
```

### Question 2
2.1. Customer Engagement Patterns
- Does tenure affect number of products, activity, or credit card ownership?

```python
# Tenure vs. Number of Products
sb.barplot(x='Tenure', y='NumOfProducts', data=comb_df, estimator='mean')
plt.title("Avg Number of Products by Tenure")
plt.xticks(rotation=45)
plt.show()

# Stacked bar: IsActiveMember by tenure
tenure_activity = pd.crosstab(comb_df['Tenure'], comb_df['IsActiveMember'], normalize='index')
tenure_activity.plot(kind='bar', stacked=True)
plt.title("Active Members by Tenure")
plt.ylabel("Proportion")
plt.legend(title='IsActiveMember', labels=['No', 'Yes'])
plt.show()

# Tenure vs. Balance
sb.lineplot(x='Tenure', y='Balance', data=comb_df)
plt.title("Average Balance by Tenure")
plt.show()
```

```python
comb_df.groupby('Tenure')['NumOfProducts'].mean().sort_values(ascending=False)
```

```python
comb_df.groupby('Tenure')['Balance'].mean().sort_values(ascending=False)
```

#### The average number of Products in Tenure 2 is high the rest of the tenure
#### The most Active members can be found in Tenure 1
#### The average balance by tenure can be found in Tenure 0

###  4. Exited Customers Analysis
   Questions
- What characteristics are common among exited customers?

```python
# Exit rate by gender
sb.barplot(x='Gender', y='Exited', data=comb_df)
plt.title("Exit Rate by Gender")
plt.show()

# Exit rate by age group (create age buckets)
comb_df['AgeGroup'] = pd.cut(comb_df['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '31-40', '41-50', '51-60', '60+'])
sb.barplot(x='AgeGroup', y='Exited', data=comb_df)
plt.title("Exit Rate by Age Group")
plt.show()

# Boxplot: balance for exited vs not
sb.boxplot(x='Exited', y='Balance', data=comb_df)
plt.title("Balance vs Exited Status")
plt.show()

# Exit rate by tenure
exit_by_tenure = comb_df.groupby('Tenure')['Exited'].mean()
exit_by_tenure.plot(kind='line', marker='o')
plt.title("Exit Rate by Tenure")
plt.ylabel("Exit Rate")
plt.show()
```

```python
# Set up visualization style
sb.set(style="whitegrid")

# Create subplots to visualize impact of key features on churn
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# Gender vs Churn
sb.countplot(data=comb_df, x='Gender', hue='Exited', ax=axes[0, 0])
axes[0, 0].set_title("Churn by Gender")

# Geography vs Churn
sb.countplot(data=comb_df, x='Geography', hue='Exited', ax=axes[0, 1])
axes[0, 1].set_title("Churn by Geography")

# Age distribution by churn
sb.boxplot(data=comb_df, x='Exited', y='Age', ax=axes[1, 0])
axes[1, 0].set_title("Age Distribution by Churn")

# Balance distribution by churn
sb.boxplot(data=comb_df, x='Exited', y='Balance', ax=axes[1, 1])
axes[1, 1].set_title("Balance Distribution by Churn")

# Number of Products vs Churn
sb.countplot(data=comb_df, x='NumOfProducts', hue='Exited', ax=axes[2, 0])
axes[2, 0].set_title("Churn by Number of Products")

# Active Member vs Churn
sb.countplot(data=comb_df, x='IsActiveMember', hue='Exited', ax=axes[2, 1])
axes[2, 1].set_title("Churn by Activity Status")
axes[2, 1].set_xticklabels(['Inactive', 'Active'])

plt.tight_layout()
plt.show()
```
